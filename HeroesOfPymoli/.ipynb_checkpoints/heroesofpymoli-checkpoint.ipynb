```python
import pandas as pd
```

```python
data_file = "../../../../gt-atl-data-pt-06-2020-u-c/DataViz-Content/04-Pandas/Homework/Instructions/HeroesOfPymoli/Resources/purchase_data.csv"
```

```python
purchase_data = pd.read_csv(data_file)
```

```python
total_players = len(purchase_data["SN"].unique())
pd.DataFrame({"Total Players" : [total_players]})
```

```python
unique_items = purchase_data["Item ID"].nunique()
average_price = purchase_data["Price"].mean()
number_purchases = purchase_data["Purchase ID"].nunique()
total_rev = purchase_data["Price"].sum()
```

```python
purchase_analysis=pd.DataFrame({
    "Unique Items":[unique_items],
    "Average Price":[average_price],
    "Number Purchases":[number_purchases],
    "Total Revenue": [total_rev]})
purchase_analysis
```

```python
player_unique = purchase_data.drop_duplicates(['SN'], keep ='last')
gender_count = player_unique['Gender'].value_counts()
gender_percent = (gender_count/total_players)*100
```

```python
  gender_df= pd.DataFrame(data=
    {"Total Count":(gender_count),
    "Percentage of Players":gender_percent})
gender_df
```

```python
gender_purchase_count=purchase_data.groupby(["Gender"])["Purchase ID"].count()
gender_avg_price=purchase_data.groupby(["Gender"])["Price"].mean()
gender_total_purchase=purchase_data.groupby(["Gender"])["Price"].sum()
gender_avg_total=gender_total_purchase/gender_count
```

```python
gender_summary_df = pd.DataFrame(data=
        {"Purchase Count":gender_purchase_count, 
        "Average Purchase Price":gender_avg_price,
        "Total Purchase Value":gender_total_purchase,
        "Average Total Purchase Per Person":gender_avg_total})
gender_summary_df
```

```python
bins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,200]
bin_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
age_unique = purchase_data.loc[:,["SN", "Age"]]
age_unique = age_unique.drop_duplicates()

age_unique["Age Ranges"] = pd.cut(age_unique["Age"], bins, labels = bin_labels)

age_totals = age_unique["Age Ranges"].value_counts()
age_percent = (age_totals/total_players)*100
```

```python
age_demo_df = pd.DataFrame(data=
    {"Total Count": age_totals, 
      "Percentage of Players": age_percent})
age_demo_df
```

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels = bin_labels)
age_group=purchase_data.groupby("Age Ranges")
age_purchase_count = age_group["Purchase ID"].count()
age_avg_price = age_group["Price"].mean()
age_total_purchase = age_group["Price"].sum()
age_avg_total = age_total_purchase/age_totals
```

```python
age_summary_df = pd.DataFrame(data=
        {"Purchase Count":age_purchase_count, 
        "Average Purchase Price":age_avg_price,
        "Total Purchase Value":age_total_purchase,
        "Average Total Purchase Per Person":age_avg_total})
age_summary_df
```

```python
spenders=purchase_data.groupby("SN")
spend_purchase_count=spenders["Purchase ID"].count()
spend_avg_price=spenders["Price"].mean()
spend_total_purchase=spenders["Price"].sum()
```

```python
spender_df = pd.DataFrame(data=
        {"Purchase Count": spend_purchase_count, 
         "Average Purchase Price": spend_avg_price, 
         "Total Purchase Value": spend_total_purchase})
```

```python
top_spender_df = spender_df.sort_values(["Total Purchase Value"], ascending=False)
top_spender_df.head()
```

```python
items_group=purchase_data.groupby(["Item ID","Item Name"])
item_purchase_count = items_group['Purchase ID'].count()
item_total_purchase = items_group["Price"].sum()
item_price=item_total_purchase/item_purchase_count
```

```python
item_df = pd.DataFrame(data=
        {"Purchase Count": item_purchase_count,
        "Item Price": item_price,
        "Total Purchase Value": item_total_purchase})
item_df = item_df.sort_values(["Purchase Count"], ascending=False)
item_df.head()
```
